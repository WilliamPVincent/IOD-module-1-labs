````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
X = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
Y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''

    npX=np.array(X)
    npY=np.array(Y)
    XYsum=np.sum(np.multiply(npX,npY))
    XsumYsum= np.multiply(np.sum(npX),np.sum(npY))
    sqrXsum= np.sum(np.square(npX))
    sumXsqr= np.square(np.sum(npX))
    N= len(X)

    m= (N*XYsum-XsumYsum)/(N*sqrXsum-sumXsqr)
    b = (np.sum(npY) - m*np.sum(npX))/N

    return m, b

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X,Y, linestyle= '-', color='red',alpha=0.5)
plt.plot(X,m*np.array(X)+b)
plt.show()
````
